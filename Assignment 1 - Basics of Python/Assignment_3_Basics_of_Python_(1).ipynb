## Exercise 1: Check Prime Number

```python

num = int(input("Enter a number: "))

if num <= 1:
    print(num, "is not a prime number")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
```


## Exercise 2: Product of Random Numbers


```python

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print("First number:", num1)
print("Second number:", num2)

user_answer = int(input("Enter the product of the two numbers: "))

if user_answer == num1 * num2:
    print("Correct answer!")
else:
    print("Incorrect. The correct product is:", num1 * num2)
```

```python
# Hospital Billing System

room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300
discount_rate = 0.10


days_stayed = 4
number_of_tests = 3
medicine_charges = 2400


room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_cost * number_of_tests


total_bill = room_charges + doctor_fee + lab_charges + medicine_charges


if total_bill > 10000:
    discount = total_bill * discount_rate
else:
    discount = 0


final_bill = total_bill - discount

# Display the bill details
print("Hospital Bill Details")
print("----------------------")
print("Room Charges       : ", room_charges)
print("Doctor Fee         : ", doctor_fee)
print("Lab Test Charges   : ", lab_charges)
print("Medicine Charges   : ", medicine_charges)
print("Total Bill         : ", total_bill)
print("Discount Applied   : ", discount)
print("Final Amount Payable:", final_bill)
```
