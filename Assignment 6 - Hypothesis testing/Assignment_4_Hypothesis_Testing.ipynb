
# Assignment 4: Hypothesis Testing

```python

# Given values
n = 25
x_bar = 3050  # sample mean weekly cost

mu_X = 600    # mean units produced (given)
sigma_X = 25  # std dev of units produced (given)

# Cost model: W = 1000 + 5X
# Theoretical mean cost (mu0) when X mean = 600
mu0 = 1000 + 5 * mu_X
mu0
```

```python

import math
sigma_W = 5 * sigma_X
sigma_W

z_stat = (x_bar - mu0) / (sigma_W / math.sqrt(n))
z_stat
```

```python
import numpy as np

sigma_W_np = 5 * sigma_X

z_stat_np = (x_bar - mu0) / (sigma_W_np / np.sqrt(n))

print(f"Sigma_W (NumPy): {sigma_W_np}")
print(f"Z-statistic (NumPy): {z_stat_np}")
```

```python

from scipy import stats

alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)  # right-tail critical value
z_critical
```

- If **z_stat > z_critical**, reject H0  
- Otherwise, fail to reject H0


```python

if z_stat > z_critical:
    decision = "Reject H0"
    conclusion = ("There is strong evidence (at 5% level) that the mean weekly operating cost "
                  "is higher than what the model suggests.")
else:
    decision = "Fail to reject H0"
    conclusion = ("There is not enough evidence (at 5% level) to say that the mean weekly operating cost "
                  "is higher than what the model suggests.")

decision, conclusion
```
