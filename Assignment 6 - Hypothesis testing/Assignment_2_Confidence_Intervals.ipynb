
# Assignment 2: Estimation and Confidence Intervals

## Objective
To construct 99% confidence intervals for the mean durability of print-heads using:
1. Sample standard deviation (t-distribution)
2. Known population standard deviation (z-distribution)



## Given Data
Durability of print-heads (in million characters before failure):


```

import numpy as np
import scipy.stats as stats

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n = len(data)
mean = np.mean(data)
std_sample = np.std(data, ddof=1)

mean, std_sample, n
```


## Part (a): 99% Confidence Interval Using Sample Standard Deviation

### Why t-distribution?
- Sample size is small (n < 30)
- Population standard deviation is unknown
- Therefore, Student's t-distribution is appropriate


```

confidence_level = 0.99
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_error_t = t_critical * (std_sample / np.sqrt(n))

ci_lower_t = mean - margin_error_t
ci_upper_t = mean + margin_error_t

t_critical, ci_lower_t, ci_upper_t
```


### Interpretation
We are 99% confident that the true mean durability of the print-heads lies within this interval.



## Part (b): 99% Confidence Interval Using Known Population Standard Deviation

Given:
- Population standard deviation = 0.2 million characters
- z-distribution is applicable


```

population_std = 0.2

z_critical = stats.norm.ppf(1 - alpha/2)
margin_error_z = z_critical * (population_std / np.sqrt(n))

ci_lower_z = mean - margin_error_z
ci_upper_z = mean + margin_error_z

z_critical, ci_lower_z, ci_upper_z
```


### Interpretation
Using the known population standard deviation, the confidence interval is narrower,
indicating greater certainty in the estimate of the mean.



## Conclusion
- t-distribution is used when population standard deviation is unknown and sample size is small.
- z-distribution is used when population standard deviation is known.
- Confidence intervals provide a range of plausible values for the population mean.
